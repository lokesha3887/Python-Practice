```python
import pandas as pd
from statistics import mode
data=pd.read_csv("weather.csv")
data
```

```python
data.info()
```

```python
print(data.Sunshine.mean())
print(data.WindGustDir.mode())
print(data.WindGustSpeed.mean())
print(data.WindDir9am.mode())
print(data.WindDir3pm.mode())
print(data.WindSpeed9am.mean())
```

```python
#filling null values by mean(for numerical data) and mode (for categorical data)
data1=data.fillna({
    'Sunshine':7.909366391184575,
    'WindGustDir':'NW',
    'WindGustSpeed':39.84065934065934,
    'WindDir9am':'SE',
    'WindDir3pm':'NW',
    'WindSpeed9am':9.651810584958218
})
data1
```

```python
data1.info()
```

```python
#forward fill -->fills null value by its previous value
data2=data.fillna(method='ffill');data2
```

```python
#backrward fill -->fills null value by its next value
data3=data.fillna(method='bfill')
data3
```

```python
#dropping rows that contains null values
data4=data.dropna()
data4
```

```python
data4.shape #number of rows reduced to 328 from 366
```

```python
#if row contain only one null value and we wish to retain that row instead of deleting the records
data5=data.dropna(thresh=1)
data5 
```

```python
data5.shape #since no row  contain more than one null records
```

```python
data=pd.read_excel("weather_data.xlsx")
data   #1000 showing error values
```

```python
#we want to replace null value by NaN
data1=data.replace(1000,np.NaN)
data1
```
