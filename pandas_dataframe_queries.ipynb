```python
#joining columns between two data sets
df1={
    'city':['banglore','mysore','chennai','hyderbad'],
    'temp':[35,30,31,32]
}
df1
d1=pd.DataFrame(df1)
d1
```

```python
d2=pd.DataFrame({
    'city':['banglore','mumbai','delhi','hyderbad'],
    'humidity':[65,60,68,63]
})
d2
```

```python
j_data=pd.merge(d1,d2,on='city')  #inner join
j_data
```

```python
j_data=pd.merge(d1,d2,on='city',how='outer')  #outer join
j_data
```

```python
j_data=pd.merge(d1,d2,on='city',how='left')  #left join
j_data
```

```python
j_data=pd.merge(d1,d2,on='city',how='right')  #right join
j_data
```

```python
j_data=pd.merge(d1,d2,on='city',how='outer',indicator=True)  #indicator 
j_data
```

```python
j_data=pd.merge(d1,d2,on='city',how='right',indicator=True) 
j_data
```

```python
#summarization using pivot table
data=pd.read_excel("SampleSuperstore.xls")
data
```

```python
#total sum of sale for each segment corresponding to cities
sumpvt=data.pivot_table(index='Segment',columns='City',values='Sales',aggfunc='sum')
sumpvt
```

```python
sumpvt=data.pivot_table(index='Segment',columns='City',aggfunc='sum')
sumpvt
```

```python
#total sales for each segment
sumpvt=data.pivot_table(index='Segment',values='Sales',aggfunc='sum')
sumpvt
```

```python
#mean sale for each segment
meanpvt=data.pivot_table(index='Segment',values='Sales',aggfunc='mean')
meanpvt
```

```python
#max sales for each segment
maxpvt=data.pivot_table(index='Segment',values='Sales',aggfunc='max')
maxpvt
```

```python
#total sales for each segment
sumpvt=data.pivot_table(index='Segment',columns='Ship Mode',margins=True)
sumpvt
```

```python
#crosstab -->contigency table
d_2017=pd.read_excel("city_weather_2017.xlsx")
d_2017
contigency_table=pd.crosstab(d_2017.city,d_2017.event)
contigency_table
```

```python
contigency_table=pd.crosstab(d_2017.city,d_2017.event,margins=True)
contigency_table
```

```python
contigency_table=pd.crosstab(d_2017.city,d_2017.event,normalize='index') #gives % based on rowsum
contigency_table
```
