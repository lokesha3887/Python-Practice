# matplot



```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
#line graph
x=[1,2,3,4,5]
y=[12,15,19,21,25]
plt.plot(x,y,marker='*')
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='dashed',marker='*')
plt.xlabel('days')
plt.ylabel('money spent')
plt.grid() #grid lines
```

```python
#bar plot
company=['google','facebook','twitter','microsoft','amazon']
revenue=[90,38,42,98,85]
profit=[38,21,18,47,41]
plt.bar(company,revenue)
```

```python
xpos=np.array([1,2,3,4,5])
xpos           
```

```python
#side by side chart
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
#histogram
x=[10,12,10,13,15,14,18,20,7,21,23,25,9,30]
plt.hist(x)
```

```python
plt.hist(x,bins=3,rwidth=0.3)
```

```python
#pie chart
plt.pie(revenue,labels=company)
```

```python
plt.pie(revenue,labels=company,radius=3,autopct='%0.3f%%',explode=[0,0,0,0,1])
plt.suptitle('Revenue of companys',size=15,y=1.5)
```

# seaborn


```python
import seaborn as sns
```

```python
data=sns.load_dataset('tips');data   #inbuilt data set
```

```python
#univariate plots
#1 strip plot
sns.stripplot(y='tip',data=data)
```

```python

sns.stripplot(x='day',y='tip',data=data)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=data,size=2,jitter=False) #jitter -->allocates points on a single line
```

```python
#swarn plot
sns.swarmplot(x='day',y='tip',data=data)  #exact distribution that is how many has given tips 
```

```python
sns.swarmplot(x='day',y='tip',data=data,hue='sex')  #hue-->for distinguish based on other var
```

```python
sns.boxplot(x='day',y='tip',data=data)
```

```python
sns.violinplot(x='day',y='tip',data=data)
```

```python
#subplot
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=data)
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=data)
```

```python
#combining plots
sns.violinplot(x='day',y='tip',data=data,inner=None,color='lightblue')
sns.stripplot(x='day',y='tip',data=data,size=3,jitter=True)
```

```python
#bivariate plots
sns.jointplot(x='total_bill',y='tip',data=data)
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde') #kernal density estimation
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,hue='sex')
```

```python
sns.boxplot(x='sex',y='tip',data=data)
```

```python
sns.boxplot(x='total_bill',y='time',hue='sex',data=data)
```
