```python
#we upload files from clipboard,html,orc,feather,hdf,json....
#using python dictionary
weather_data={
    'day': ['1/1/2001','2/1/2001','3/1/2001','4/2/2001'],
    'temperature': [32,33,31,36],
    'windspeed': [6,5,7,5],
    'event': ['rain','sunny','snow','rain']
}    
weather_data
data=pd.DataFrame(weather_data)
data
```

```python
#using list tuples
weather_data=[('1/1/2001',32,6,'rain'),
             ('2/1/2001',33,5,'sunny'),
             ('3/1/2001',31,7,'snow'),
             ('4/1/2001',36,5,'rain')
             ]
data=pd.DataFrame(weather_data,columns=["day","temperature","windspeed","event"])
data
```

```python
#using list of dictionaries
weather_data=[{'day':'1/1/2001','temperature':32,'windspeed':6,'event':'rain'},
             {'day':'2/1/2001','temperature':35,'windspeed':5,'event':'sunny'},
             {'day':'3/1/2001','temperature':36,'windspeed':7,'event':'rain'}
             ]
data=pd.DataFrame(weather_data)
data
```
